$f(x)=1260+117x-9x^2$

```python
import sympy as sp
import matplotlib.pyplot as plt
%matplotlib inline
sp.var("x")

def f(x) :
    return (1260+(117*x)-(9*x**2))
df = sp.diff (f(x),x)
pc = sp.solve(df,x)
```

```python
pc
```

```python
import numpy as np

xn = np.linspace(start = 1, stop = 7, num = 50)
yn = f(xn)
# duda 
```

```python
plt.figure()
plt.plot(xn,yn)
plt.grid()
```
